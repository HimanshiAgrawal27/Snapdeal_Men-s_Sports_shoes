```python
import requests
```

```python
url="https://www.snapdeal.com/search?clickSrc=top_searches&keyword=mens%20sport%20shoes&categoryId=0&vertical=p&noOfResults=20&SRPID=topsearch&sort=rlvncy&q=Brand%3Aaadiva%5EAirhype%5EASE%20PRO%20SPORT%5EBirde%5EBolinzer%7C"
headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36"}

k=requests.get(url,headers=headers)
k
```

```python
from bs4 import BeautifulSoup as b1
```

```python
b=b1(k.text,"html.parser")
print(b.prettify())
```

```python
cn=b.find_all("p",class_="product-title")
brand=[]
for i in cn:
    text=i.text.strip().split()[0]
    brand.append(text)
print("Brand name:",brand)

pr=b.find_all("p",class_="product-title")
name=[]
for i in pr:
    text= " ".join(i.text.strip().split()[1:])
    name.append(text)
print("\nProduct name:",name)   

ori_price=b.find_all("span",class_="lfloat product-desc-price strike")
price=[]
for i in ori_price:
    t=i.text
    price.append(t)
print("\nOriginal price:",price) 

disc_price=b.find_all("span",class_="lfloat product-price")
discprice=[]
for i in disc_price:
    g=i.text
    discprice.append(g)
print("\n Discount price:",discprice)  

dis_per=b.find_all("div",class_="product-discount")
disp=[]
for i in dis_per:
    l=i.text.replace('\n','').replace('Off','').strip()
    disp.append(l)
print("\nDiscount percentage:",disp) 

rating_tags = b.find_all("div", class_="filled-stars")
review_tags = b.find_all("p", class_="product-rating-count")

rating = []
reviews = []

for r in rating_tags:
    if "style" in r.attrs:
        star = float(r["style"].replace("width:", "").replace("%", "")) / 20
        rating.append(star)
    else:
        rating.append("Not Rated")
for rv in review_tags:
    count = rv.text.strip().replace('(','').replace(')','')  
    reviews.append(count)

while len(reviews) < len(brand):
    reviews.append("0")

while len(rating) < len(brand):
    rating.append("Not Rated")

print("\nRating:", len(rating))
print("\nReviews:", reviews)       



```

```python
import pandas as pd
```

```python
df1=pd.DataFrame({"Product_Name:":name,"Brand":brand,"Original_Price":price,"Discount_price":discprice,"Discounted_percentage":disp,"Rating":rating,"Reviews":reviews})
df1.to_csv("Men's_sports_shoes_datasettt.csv")
print("File saved successfully")
df1
```
